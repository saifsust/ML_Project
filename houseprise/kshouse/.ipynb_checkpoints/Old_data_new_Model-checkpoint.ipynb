# Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split

reg = LinearRegression()
labels = data['price']
train1 = data.drop(['price'],axis=1)


x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
# train1.describe()
# print(x_train)
# print(y_train)
reg.fit(x_train,y_train)
```

```python
acc=reg.score(x_test,y_test)
print("Accuracy %.2f" % round(acc*100,2)+' %')
```

# Gradient Boosting

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.08, loss = 'ls')

clf.fit(x_train, y_train)
```

```python
acc=clf.score(x_test,y_test)
print("Accuracy %.2f" % round(acc*100,2)+' %')
```

```python
t_sc = np.zeros(400,dtype=np.float64)
y_pred = reg.predict(x_test)


for i,y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]=clf.loss_(y_test,y_pred)


testsc = np.arange(400)+1

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```

```python
predicted_price = clf.predict(x_test)  

predicted_price = pd.DataFrame(predicted_price,index=y_test.index,columns = ['price'])  

predicted_price.plot(figsize=(10,5))  

y_test.plot()  

plt.legend(['predicted_price','actual_price'])  

plt.ylabel("Gold ETF Price")  

plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=8, max_depth=8, random_state=12, verbose=0)

rfr.fit(x_train, y_train)
```

```python
acc=rfr.score(x_test,y_test)
print("Accuracy %.2f" % round(acc*100,2)+' %')
```

```python
predicted_price = rfr.predict(x_test)  

predicted_price = pd.DataFrame(predicted_price,index=y_test.index,columns = ['price'])  

predicted_price.plot(figsize=(10,5))  

y_test.plot()  

plt.legend(['predicted_price','actual_price'])  

plt.ylabel("Gold ETF Price")  

plt.show()
```

# Lasso Regression

```python
from sklearn import linear_model
lrg = linear_model.Lasso(alpha = 0.1)
lrg.fit(x_train,y_train)
```

```python
acc=lrg.score(x_test,y_test)
print("Accuracy %.2f" % round(acc*100,2)+' %')
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
ridgereg = Ridge(0.08,normalize=True)
ridgereg.fit(x_train,y_train)
acc=ridgereg.score(x_test,y_test)
print("Accuracy %.2f" % round(acc*100,2)+' %')
```

```python
predicted_price = ridgereg.predict(x_test)  

predicted_price = pd.DataFrame(predicted_price,index=y_test.index,columns = ['price'])  

predicted_price.plot(figsize=(10,5))  

y_test.plot()  

plt.legend(['predicted_price','actual_price'])  

plt.ylabel("Gold ETF Price")  

plt.show()
```
